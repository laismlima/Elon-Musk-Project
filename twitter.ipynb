```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python

# Setting variables to be used below
maxTweets = 1000

# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:elonmusk').get_items()):
    if i>maxTweets:
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.favorite_count, tweet.content, tweet.username])
```

```python
# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Like', 'Text', 'Username'])

# Display first 5 entries from dataframe
tweets_df1.head(20)
```

```python
# Export dataframe into a CSV

tweets_df1.to_csv(r'D:\Desktop\Elon3.csv', sep = ';', index=False)
```
